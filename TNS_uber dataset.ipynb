```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import easygui
```

```python
# Step 1: Data Collection and Loading
```

```python
path= easygui.fileopenbox()    #Load Dataset
```

```python
df=pd.read_csv(path)            
df
```

```python
df.head()     # Display first rows
```

```python
print("Rows, Columns:", df.shape)            # Rows and Columns
```

```python
# Step 2: Understanding Dataset
```

```python
df.columns        # Column names
```

```python
df.dtypes #Data Types
```

```python
df.nunique()  #Unique values
```

```python
# Step 3: Data Preprocessing
```

```python
df.isnull().sum()          #Handle missing values
```

```python
df.dropna()   # remove missing values
```

```python
df.drop_duplicates()      #Remove Duplicates
```

```python
df.columns=df.columns.str.lower().str.replace("_","")             #Rename Columns
df.columns
```

```python
df
```

```python
# Step 4: Exploratory Data Analysis
```

```python
df.describe() #Analyze distribution
```

```python
df.info() #Analyze distribution
```

```python
df
```

```python
category_miles=df.groupby('category')['miles'].mean()   #Group Data
category_miles
```

```python
purpose_miles=df.groupby('purpose')['miles'].mean()    #Group Data
purpose_miles
```

```python
# Data Visulization
```

```python
df
```

sns.histplot(data=df, x='purpose',kde=True)    #Histogram

```python
sns.barplot(data=df,x='miles', y='purpose')   #Bar chart
```

```python
sns.scatterplot(data=df, x='category', y='miles')   #scatter plot
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)    #Heat map
```
