```python
# FINAL OPTIMIZED ENTERTAINMENT CHATBOT

import random
import nltk
import string
import numpy as np

from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Download NLTK DATA
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# KNOWLEDGE BASE
intents = {
    "intents": [
        {
            "tag": "greeting",
            "patterns": ["hi", "hello", "hey", "what's up"],
            "responses": ["Hey! Ready for some fun? 😄", "Hello! Let's enjoy 🎉"]
        },
        {
            "tag": "jokes",
            "patterns": ["joke", "tell me a joke", "funny"],
            "responses": ["Why don’t programmers like nature? Too many bugs! 😂"]
        },
        {
            "tag": "movies",
            "patterns": ["movie", "suggest movie"],
            "responses": ["Try *Inception* 🎬"]
        },
        {
            "tag": "music",
            "patterns": ["music", "songs"],
            "responses": ["Try lo-fi beats 🎧"]
        },
        {
            "tag": "game",
            "patterns": ["play game", "game"],
            "responses": ["Let's play a number guessing game!"]
        },
        {
            "tag": "goodbye",
            "patterns": ["bye", "exit"],
            "responses": ["Goodbye! 😎"]
        }
    ]
}

# PREPROCESSING
def preprocessor(text):
    text = text.lower().strip()
    text = text.replace('"', '').replace("'", "")

    tokens = nltk.word_tokenize(text)

    tokens = [
        lemmatizer.lemmatize(word)
        for word in tokens
        if word not in string.punctuation and word not in stop_words
    ]

    return " ".join(tokens)

# TRAINING DATA
corpus = []
labels = []

for intent in intents["intents"]:
    for pattern in intent["patterns"]:
        corpus.append(preprocessor(pattern))
        labels.append(intent["tag"])

# VECTORIZATION
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(corpus)

# MODEL
model = LogisticRegression(max_iter=200)
model.fit(X, labels)

# CONTEXT MEMORY
context = {
    "last_intent": None,
    "game_number": None
}

# CHATBOT FUNCTION
def chatbot_response(user_input):
    processed = preprocessor(user_input)
    vector = vectorizer.transform([processed])

    probs = model.predict_proba(vector)[0]
    confidence = np.max(probs)
    intent_tag = model.predict(vector)[0]

    print(f"[DEBUG] Intent: {intent_tag}, Confidence: {confidence:.2f}")

    # LOW CONFIDENCE
    if confidence < 0.25:
        return random.choice([
            "Not sure I got that, but I can tell jokes, suggest movies or music!",
            "Try asking for jokes, movies, or games",
            "I'm still learning! Ask me something fun"
        ])

    # GAME START
    if intent_tag == "game":
        context["game_number"] = random.randint(1, 5)
        context["last_intent"] = "game"
        return "I picked a number between 1 and 5. Guess it!"

    # GAME CONTINUE
    if context["last_intent"] == "game" and user_input.isdigit():
        guess = int(user_input)
        if guess == context["game_number"]:
            context["last_intent"] = None
            return "Correct! You win!"
        else:
            return "Nope! Try again."

    context["last_intent"] = intent_tag

    for intent in intents["intents"]:
        if intent["tag"] == intent_tag:
            return random.choice(intent["responses"])

# CHAT LOOP
print("Entertainment AI Chatbot is Live! (type 'exit' to stop)")

while True:
    user = input("You: ")

    if user.lower() in ["exit", "quit"]:
        print("Bot: Bye! Stay happy!")
        break

    print("Bot:", chatbot_response(user))
```

```python
def data_analyst_bot():
  data = input("Paste dataset (table or text):\n")

  prompt = f"""
  Analyze the data and provide:
  - Key insights
  - Trends
  - Annomalies
  - Business recommendations
  """

  # USE THE CHATBOT_RESPONSE FUNCTION FOR RETRY LOGIC AND GEMINI-PRO MODEL
  response_text = chatbot_response(prompt + data)
  print("\nInsights:\n", response_text)

data_analyst_bot()
```
