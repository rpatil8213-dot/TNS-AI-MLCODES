```python
df = pd.DataFrame(xtrain.reshape(60000,-1))  #converting img to the dataframe
df
```

```python
plt.imshow(df.iloc[0,:].values.reshape(28,28))  ## Dataframe to img show
```

```python
model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Dense(8, activation = 'sigmoid', input_dim = (784)),
        tf.keras.layers.Dense(10, activation = 'softmax' )
    ])
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
ytrain_encod = pd.get_dummies(ytrain)
```

```python
history = model.fit(df.values, ytrain_encod, 
                    batch_size = 5, epochs = 5,
                    validation_data = (xtest.reshape(10000,-1), pd.get_dummies(ytest)))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
