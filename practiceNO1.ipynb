```python
#41. Find all pairs in a list whose sum equals agiven number.
numbers = [1, 2, 3, 4, 5, 6]
target = 7

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] + numbers[j] == target:
            print(numbers[i], numbers[j])
```

```python
numbers = [1, 2, 3, 4, 5, 6]
target = 7

seen = set()

for num in numbers:
    complement = target - num
    if complement in seen:
        print(complement, num)
    seen.add(num)
```

```python
#42.Flatten a nested list (example: [[1,2],[3,4]] → [1,2,3,4]).
nested_list = [[1, 2], [3, 4]]

flattened = [num for sublist in nested_list for num in sublist]

print(flattened)
```

```python
nested_list = [[1, 2], [3, 4]]

flattened = []

for sublist in nested_list:
    for num in sublist:
        flattened.append(num)

print(flattened)
```

```python
import itertools

nested_list = [[1, 2], [3, 4]]

flattened = list(itertools.chain(*nested_list))

print(flattened)
```

```python
#43 find duplicate element in a list.
numbers = [1, 2, 3, 4, 2, 5, 3, 6]

seen = set()
duplicates = set()

for num in numbers:
    if num in seen:
        duplicates.add(num)
    else:
        seen.add(num)

print(duplicates)
```

```python
#56. Group words by their length using a dictionary.
words = ["cat", "dog", "elephant", "bat", "lion", "tiger"]

grouped = {}

for word in words:
    length = len(word)
    
    if length not in grouped:
        grouped[length] = []
        
    grouped[length].append(word)

print(grouped)
```

```python
#60. From a list of number , create:
# a list of even numbers
# a tuple of odd numbers
# a set of unique numbers
# a dictionary with number and its square


numbers = [1, 2, 3, 4, 5, 6, 2, 3]

# List of even numbers
even_list = [num for num in numbers if num % 2 == 0]

# Tuple of odd numbers
odd_tuple = tuple(num for num in numbers if num % 2 != 0)

# Set of unique numbers
unique_set = set(numbers)

# Dictionary with number and its square
square_dict = {num: num**2 for num in numbers}

print("Even list:", even_list)
print("Odd tuple:", odd_tuple)
print("Unique set:", unique_set)
print("Square dictionary:", square_dict)
```
