## Common Functions Quick Reference

**Data Exploration:**
- `df.head(n)` - First n rows
- `df.tail(n)` - Last n rows
- `df.shape` - Dimensions (rows, columns)
- `df.info()` - Data types and memory info
- `df.describe()` - Statistical summary
- `df.columns` - Column names
- `df.dtypes` - Data types of columns

**Data Cleaning:**
- `df.isnull().sum()` - Count missing values
- `df.fillna(value)` - Fill missing values
- `df.dropna()` - Drop rows with missing values
- `df.drop_duplicates()` - Remove duplicate rows
- `df.duplicated().sum()` - Count duplicates

**Data Analysis:**
- `df['col'].value_counts()` - Count unique values
- `df['col'].nunique()` - Count of unique values
- `df['col'].unique()` - List unique values
- `df.groupby('col')` - Group data
- `df.sort_values('col')` - Sort data
- `df.loc[]` - Select by label
- `df.iloc[]` - Select by position
- `df.nlargest(n, 'col')` - Top n values
- `df.nsmallest(n, 'col')` - Bottom n values

**String Operations:**
- `df['col'].str.lower()` - Convert to lowercase
- `df['col'].str.upper()` - Convert to uppercase
- `df['col'].str.contains('text')` - Check if contains text
- `df['col'].str.split()` - Split string
- `df['col'].str.len()` - String length

**Statistical Functions:**
- `df['col'].mean()` - Average
- `df['col'].median()` - Median
- `df['col'].mode()` - Mode
- `df['col'].std()` - Standard deviation
- `df['col'].var()` - Variance
- `df['col'].quantile(0.25)` - Percentile
- `df['col'].min()` - Minimum
- `df['col'].max()` - Maximum

**Aggregation:**
- `df.groupby('col').mean()` - Group and calculate mean
- `df.groupby('col').agg({'col1':'mean', 'col2':'sum'})` - Multiple aggregations
- `pd.pivot_table(df, values, index, columns, aggfunc)` - Pivot table
- `pd.crosstab(df['col1'], df['col2'])` - Cross tabulation

**Visualization:**
- `plt.plot()` - Line plot
- `plt.hist()` - Histogram
- `plt.bar()` - Bar chart
- `plt.scatter()` - Scatter plot
- `plt.pie()` - Pie chart
- `plt.subplots(rows, cols)` - Multiple plots
- `sns.boxplot()` - Box plot
- `sns.violinplot()` - Violin plot
- `sns.heatmap()` - Heatmap
- `sns.countplot()` - Count plot

**Conditional Operations:**
- `np.where(condition, value_if_true, value_if_false)` - Conditional assignment
- `pd.cut(series, bins, labels)` - Bin continuous data
- Boolean indexing: `df[df['col'] > value]`
- Multiple conditions: `df[(condition1) & (condition2)]`
