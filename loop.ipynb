```python
numbers = [11, 12, 13, 14, 15]
for i in numbers:
    print(i**2)
```

```python
# for loop to iterate over a list
numbers = [12, 13, 14, 16, 17]
for i,j in enumerate(numbers):
    print(i,j**2)
```

```python
#Example 1- Iterating Over a list
# list of fruits
fruits = ["apple", "banana", "cherry", "dates"]

# for loop to iterate over the list
for i in fruits:
    print(i)
```

```python
#example-2 
#string to iterate over
word = "hello world."

# for loop to iterte over the string
for i,j in enumerate(word):
    print(i,j)
```

```python
#example-4 iterating over a set
#set of colors
colors = {"red", "green", "blue"}
#for loop to iterate over the set
for color in colors:
    print(color)
```

```python
#example-5 iterating over a tuple
# tuple of co-ordinates
coordinates = (10, 20, 30)
# for loop to iterate over the tuple
for i in coordinates:
    print(i)
```

```python
# WHILE LOOP
# while loop to print number from 1 to 5
count = 1
while count <= 5: # when it true untill run
    print(count)
    count = count+ 1
```

```python
for i in range (1,51):  #  range(star_num, end_num+1)
    if i%5==0 and i%3==0:
        print(i, "frizbuzz")
    elif i%3==0:
        print(i, 'fizz')
    elif i%5==0:
        print(i, 'buzz')
    else:
        print(i)
    
```
