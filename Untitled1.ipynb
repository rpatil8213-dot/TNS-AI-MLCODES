```python
# Step 1: Import Required Libraries

import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

#Download required resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
#Step 2: Sample Text
text= "Natural Language Processing is amazing! It helps machines understand human language"
```

```python
#Step 3: Text Cleaning- (Lowercase+ Remove Special characters)
clean_text = re.sub(r'[^a-zA-Z]',' ', text).lower()
clean_text
```

```python
#Step 4: Tokenization (Split text into words)
tokens = word_tokenize(clean_text)
tokens
```

```python
#Step 5: Stopword Removal (Remove common words like is, the, and)
stop_words=set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]
filtered_tokens
```

```python
#Step 6: Stemming (Reduce words to root form)
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
stemmed_words
```

```python
#Lemmatization (Convert words to meaningful base from dictionary)
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
lemmatized_words
```

```python
#Output
print('Original Text:', text)
print('Clean Text:', clean_text)
print('Tokens', tokens)
print('After Stopword Removal:', filtered_tokens)
print('Stemmed Words:', stemmed_words)
print('Lemmatized Words:', lemmatized_words)
```

```python
import re
```

```python
text1 = "WOW!!! NLP in 2025 is sooo COOOOL 😍😍"
text2 = "I can't believe this!!! It's 100% true..."
text3 = "This     sentence     has      extra spaces!!!"
text4 = "Nooooo wayyyyy this works????"

texts = [text1, text2, text3, text4]


# Step 2: Contraction Expansion
def expand_contractions(text):
  contractions = {
      "can't": "cannot",
      "it's" : "it is",
      "don't": "do not",
      "won't": "will not"
  }

  for c, e in contractions.items():
    text = re.sub(r"\b"+ c + r"\b", e, text)
  return text


# Step 3: Text Cleaning Function
def clean_text(text):
  text = text.lower()
  text = expand_contractions(text)
  text = re.sub(r'\d+', '', text)
  text = re.sub(r'[^a-z\s]', '', text)
  text = re.sub(r'(.)\1{2,}', r'\1\1', text)
  text = re.sub(r'\s+', " ", text).strip()
  return text


# Step 4: Apply Cleaning
print("TEXT CLEANING OUTPUT:\n")

for text in texts:
  cleaned = clean_text(text)
  print("Original :", text)
  print("Cleaned :", cleaned)
  print("-" * 50)
```
