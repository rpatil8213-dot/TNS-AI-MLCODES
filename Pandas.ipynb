```python
import pandas as pd

data = [
    {'Name': 'Aman', 'Age': 24, 'City': 'Pune', 'gender': 'Male'},
    {'Name': 'Bob', 'Age': 25, 'City': 'Kochi'},
    {'Name': 'Chandu', 'Age': 35, 'City': 'Chennai'}
]

df = pd.DataFrame(data)

df
```

```python
#  STEP 2 data cleaning adv
df["Name"] =  df["Name"].fillna("Unknown")
df["gender"] = df["gender"].fillna("Not Specified")

df["Name"] = df["Name"].str.lower()
df["City"] = df["City"].str.upper()
df["gender"] = df["gender"].str.capitalize()
display(df)
```

```python
# step 3 conditional colum (lamdba)
df["Age_Group"] = df["Age"].apply(
    lambda x: "Young" if x < 30 else "Mid" if x < 35 else "Senior"
)

print(df)
```

```python
#step 4 map vs apply
age_map = {25: "Junior", 30:"Associate", 35: "Senior"}
df["Experincde_Level"] = df["Age"].map(age_map)
display(df)
```

```python
#STEP 5. vectorize multiple cindition
conditions = [
    (df["Age"] < 30),
    (df["Age"].between (30,34)),
    (df["Age"] >= 35)
]
choices = ["Low", "Medium", "High"]
df["Risk_Level"] = np.select(conditions, choices, default="Unknown")
display(df)
```

```python
# STEP6 query(SQL like filtering) filtering the data frame to create filtered_df
filtered_df = df.query("Age >= 30 and City != 'Pune'")
display(filtered_df)
```

```python
#step 7 assign (chain operation) adding 'Age_Squared' and 'Name_length' colums using chaianed assign operation

df = (
    df.assign(
        age_Squared = lambda x: x["Age"] ** 2,
        Name_Length = lambda x: x["Name"].str.len()
    )
)
display(df)
```

```python
#step 8  Apply Across Rows (Axis=1)
#Creating aPrifile column by applying a function row Wise
df["Profile"] = df.apply(
    lambda row: f"{row['Name']} ({row['City']}) - {row['Age']} yrs old",
    axis=1
)

display(df)
```

```python
#step=9 String advanced operations
df["City_Code"] = df["City"].str[:3]
df["Name_Initial"] = df["Name"].str[0]
display(df)
```

```python
#Step=10 Sort + Rank
df["Age_Rank"] = df["Age"].rank(ascending=False)
df = df.sort_values("Age_Rank")
print(df)
```

```python
#Step=11 Window Function(calculating the 'Age_Rolling_Mean'using a rolling window)
df['Age_Rolling_Mean'] = df['Age'].rolling(window=2).mean()
print(df)
```

```python
#step=12 Dupliucate checking (adv) identifying duplicate names in the dataframe
df["Is_Duplicate_Name"] = df.duplicated(subset=["Name"])
print(df)
```

```python
#Step=13 Memory & Performance
df["Age"] = df["Age"].astype("int16")
print(df.dtypes)
```

```python
#Step=14 Pipe (Function Style)
def normalize_age(frame):
    frame["Age_Normalized"] = (frame["Age"] - frame["Age"].min()) / (
        frame["Age"].max() - frame["Age"].min()
    )
    return frame

df = df.pipe(normalize_age)

print(df)
```

```python
#Step=15 Final Output summary
print("/nFinal Transformed DataFrame:")
print(df)
print("/nFiltered DataFrame (from earlier query):")
print(filtered_df)
print("/nData Types of Final DataFrame:")
print(df.dtypes)
```
