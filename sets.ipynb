```python
# operations on set
s1= {1,2,3,4,5,6,7}
s2= {3,4,5,6,7,8,9}
```

```python
s1.union(s2) # union marj the set but not same element
```

```python
s1|s2 # union, all element from both sets
```

```python
s1 & s2 #intersection , common element 
```

```python
s1 ^ s2 # symetric diffrence, uncommon element
```

```python
s1 - s2 #unique element from only s1 , difference
```

```python
s2 - s1 #unique element from only s2 , difference
```
