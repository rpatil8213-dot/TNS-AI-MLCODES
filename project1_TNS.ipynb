```python
#Sample data both the columns should be the numerical
x=df['hsc_p']
y=df['ssc_p']

#Creating a line plot
plt.plot(x,y, label="Line")
plt.xlabel('hsc 12th percent')
plt.ylabel('ssc 10th percent')
plt.legend()
plt.show()
```

```python
#Creating a scatter plot
x=df['ssc_p']
y=df['hsc_p']
plt.scatter(x,y,label="Scatter", color='red')
plt.xlabel('hsc 12th percent')
plt.ylabel('ssc 10th percent')
plt.legend()
plt.show()
```

```python
#Creating a bar plot
x=df['gender']
y=df['salary']
plt.bar(x,y,label="Bar", color='green')
plt.xlabel("Categories")
plt.ylabel('Values')
plt.title('Bar plot')
plt.legend
plt.show()
```

```python
#Creating a histogram
data = df['ssc_p']
plt.hist(data, bins=10, label="Histogram", color='g')
plt.xlabel("Value")
plt.ylabel('Frequency')
plt.title("Histogram")
plt.legend()
plt.show()
```
