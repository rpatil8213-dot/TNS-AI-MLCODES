```python
#1. function with default parameters

def greet(name, message="Good Morning", message1="How are you"):
   # """Function to greet a person with a given message"""
    return f"Hello, {message}, {message1}, {name}!"
```

```python
# calling the function with and without the message parameter
print(greet("Shashi", 'Good Evening')) 
```

```python
#2, Function with variable length arguments

def sum_all(x,y):
    return(x+y)
sum_all(7,34)
```

```python
def sum_all(*args): # function to return the sum of all argument
    return sum(args)
# calling the function with a variable of argument
print(sum_all(1, 2, 3))
print(sum_all(4, 5, 6, 7, 8))
```

```python
#3. function with keyword arguments
def introduce(name, age, **kwargs):

    info = f"Name: {name}, Age: {age}"
    for key, value in kwargs.items():
        info = info + f", {key.capitalize()}: {value}"
    return info

print(introduce("alice", "18"))
```

```python
#4. function returing multiple values
def get_student_info():
    name = "Alice"
    age = 20
    grade = "A"
    return name, age, grade
```
